```python
from numpy import array
```

```python
ket0 = array([1,0])
ket1 = array([0,1])
ket0/2 + ket1/2
```

```python
M1 = array([[1,1],[0,0]])
M2 = array([[1,1],[1,0]])
M1/2 + M2/2
```

```python
from numpy import matmul

display(matmul(ket0,ket1))
display(matmul(ket1,M1))
```

```python
from qiskit.quantum_info import Statevector
from numpy import sqrt
```

```python
u = Statevector([1/sqrt(2),1/sqrt(2)])
```

```python
u.draw('latex')
```

```python
u.is_valid()
```

```python
v = Statevector([(1+2.0j)/3,-2/3])
v.draw('latex')
```

```python
v.is_valid()
```

```python
u.measure()
```

```python
v.measure()
```

```python
from qiskit.visualization import plot_histogram
statistics = u.sample_counts(1000)
display(statistics)
plot_histogram(statistics)
```

```python
from qiskit.quantum_info import Operator
X = Operator ([[0,1],[1,0]])
v = Statevector([1,0])
v.draw('latex')
```

```python
v = v.evolve(X)
v.draw('latex')
```

```python
u = Statevector([sqrt(2)/sqrt(3),1/sqrt(3)])
u.is_valid()
u.draw('latex')
```

```python
u = u.evolve(X)
u.draw('latex')
```

```python
from qiskit import QuantumCircuit
circuit = QuantumCircuit(1) #1Qubit

circuit.x(0)
circuit.draw()
```

```python
u = Statevector([sqrt(2)/sqrt(3),1/sqrt(3)])
u.draw('latex')
```
